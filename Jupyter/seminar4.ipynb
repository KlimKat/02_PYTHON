## Задача 1

```python
df = pd.read_csv('laptop_price.csv', encoding="windows-1251")
df.head()
```

```python
def get_gb(memory):
    lst = memory.split()
    gb = 0
    for el in lst:
        if 'GB' in el:
            gb += float(el[:-2])
        if 'TB' in el:
            gb += float(el[:-2]) * 1000
    
    return gb

df['Memory_amount'] = df['Memory'].apply(get_gb)
df
```

```python
plt.figure(figsize=(15,6))
plt.hist(df['Memory_amount'], bins=15)
plt.title('Memory')
plt.xlabel('Memory_amount')
plt.ylabel('Quantity');
```

```python
plt.figure(figsize=(15,6))
plt.hist(df['Price_euros'], bins=15)
plt.title('Prices')
plt.xlabel('Price')
plt.ylabel('Quantity');
```

```python
df['Weight'] = df['Weight'].str.replace('kg', '').astype('float')
df
```

```python
plt.hist(df['Weight'], bins=15)
plt.title('Weights')
plt.xlabel('Weight')
plt.ylabel('Quantity');
```

```python
sns.histplot(df['Weight'], bins=15)
plt.title('Weights')
plt.xlabel('Weight')
plt.ylabel('Quantity');
```

## Задача 2

```python
def get_memory_type(memory):
        type_memory = ''
        if 'HDD' in memory:
            type_memory += 'HDD'
        if 'SSD' in memory:
            type_memory += 'SSD'
        if 'Flash Storage' in memory:
            type_memory += 'Flash Storage'
        if 'Hybrid' in memory:
            type_memory += 'Hybrid'
        return type_memory.strip()

df['Memory_Type'] = df['Memory'].apply(get_memory_type)
df.head()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df['Company'], bins=15)
plt.title('Memory type by company')
plt.xlabel('Company')
plt.ylabel('Quantity')
plt.xticks(rotation=30);
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df['OpSys'], bins=15)
plt.title('System type')
plt.xlabel('OpSys')
plt.ylabel('Quantity')
plt.xticks(rotation=30);
```

```python
data = df[df['Cpu_Company'] !='Samsung'][['Cpu_Company', 'Company']].value_counts().reset_index()
data
```

```python
data = pd.crosstab(index = df['Company'], columns=df['Cpu_Company']).reset_index()
data.drop(columns=['Samsung'], inplace=True)
data
```

```python
plt.figure(figsize=(10,6))
n_ticks = np.arange(len(data['Company']))
offset = 0.2
w = 0.4
plt.bar(n_ticks - offset, data['AMD'], width=w)
plt.bar(n_ticks + offset, data['Intel'], width=w)
plt.title('Alter')
plt.xlabel('CPU')
plt.ylabel('Qantinty')
plt.legend(['AMD', 'Intel'])
plt.xticks(n_ticks, data['Company'], rotation=50) ;
```

```python
data = pd.crosstab(index = df['Company'], columns=df['Cpu_Company'], normalize='index').reset_index()
data.drop(columns=['Samsung'], inplace=True)
data
```

```python
plt.figure(figsize=(10,6))
plt.bar(data['Company'], data['Intel'])
plt.bar(data['Company'], data['AMD'], bottom=data['Intel'])
plt.title('Alter')
plt.xlabel('CPU')
plt.ylabel('Qantinty')
plt.legend(['Intel', 'AMD'])
plt.xticks(n_ticks, data['Company'], rotation=50);
```
