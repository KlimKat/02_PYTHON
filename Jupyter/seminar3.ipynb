#### 1.1 Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU (Intel, AMD и т.д)

```python
df['Cpu_Company'] = df['Cpu'].apply(lambda x: x.split()[0])
df
```

#### 1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя (128, 1024, 256 и т.д.)

```python
def get_gb(memory):
    lst = memory.split()
    gb = 0
    for el in lst:
        if 'GB' in el:
            gb += float(el[:-2])
        if 'TB' in el:
            gb += float(el[:-2]) * 1000
    
    return gb

df['Memory_amount'] = df['Memory'].apply(get_gb)
df
```

#### 1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD, SDD и т.д.)

```python
def get_memory_type(memory):
        type_memory = ''
        if 'HDD' in memory:
            type_memory += 'HDD'
        if 'SSD' in memory:
            type_memory += 'SSD'
        if 'Flash Storage' in memory:
            type_memory += 'Flash Storage'
        if 'Hybrid' in memory:
            type_memory += 'Hybrid'
        return type_memory.strip()

df['Memory_Type'] = df['Memory'].apply(get_memory_type)
df
```

```python
df['Memory_Type'].isna().sum() #Поиск пропущенных
```

#### 1.4 Удалите признаки Memory и ScreenResolution

```python
df.drop(columns = ['Memory', 'ScreenResolution'], inplace=True)
df
```

### 5.2 С каким типом памяти и с каким объемом памяти больше ноутбуков?

```python
def getb(my_string):
    keywords = {"HDD", "SSD", "Flash", "Hybrid"}
    myset=set(my_string.split()) & keywords
    return ', '.join(myset)

df_4['Memory_Type'] = df['Memory'].apply(getb)
df_4.head()
```

```python
pd.crosstab(index=df_4['Memory_Type'], columns=df_4['Memory_amount'])
```

```python
pd.crosstab(index=df_4['Memory_amount'], columns=df_4['Memory_Type'], margins=True, margins_name='Total')
```
